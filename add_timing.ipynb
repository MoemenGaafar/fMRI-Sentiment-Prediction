```python
def remove_punctuation_and_whitespace(input_string):
    # Use a regular expression to match and remove punctuation and whitespace
    cleaned_string = re.sub(r'[^\w\s]', '', input_string).strip()
    cleaned_string = re.sub(r'«', '', cleaned_string).strip()
    return cleaned_string
```

```python
correspondences = [("don", "do"), ("im", "i"), ("hey", "they"), ("doesn", "does"), ("it ll", "it"), ("tu", "-tu")]
def are_same_words_ignore_spelling_difference(word1, word2):
    word1 = remove_punctuation_and_whitespace(word1)
    word2 = remove_punctuation_and_whitespace(word2)
#    print(word1, word2)
    if (word1, word2) in correspondences:
        return True
    #return len(word1) > 0 and word1[0] == word2[0] and word1[-1] == word2[-1] and abs(len(word1) - len(word2)) <= 2
    return len(word1) > 0 and (word1 == word2 or word1[1:] == word2)
```

```python
timing_idx = 0
timing_files_idx = 0
timings = textgrids.TextGrid(timing_files_list[timing_files_idx])[current_timings_key]

for i in range(label_df.shape[0]):
    print(i)
    sentence = label_df.iloc[i]["sentence"]
    sentence_list = sentence.split()
    print(sentence)

    # Finding the first word
    
    first_word = sentence_list[0]
    
    while timing_idx < len(timings) and not are_same_words_ignore_spelling_difference(timings[timing_idx].text, first_word):
        timing_idx += 1
        
    if timing_idx == len(timings):
        timing_files_idx += 1
        timings = textgrids.TextGrid(timing_files_list[timing_files_idx])[current_timings_key]
        timing_idx = 0
        while not are_same_words_ignore_spelling_difference(timings[timing_idx].text, first_word):
            timing_idx += 1

    t1 = timings[timing_idx].xmax

    last_word = sentence_list[-1]
    print("-------------", last_word)
    print(timing_idx, timing_files_idx)
    while not are_same_words_ignore_spelling_difference(timings[timing_idx].text, last_word):
        timing_idx += 1
        print(timings[timing_idx].text)
    
    t2 = timings[timing_idx].xmax

    label_df.at[i, "t1"] = t1
    label_df.at[i, "t2"] = t2
    label_df.at[i, "section"] = timing_files_idx
```
